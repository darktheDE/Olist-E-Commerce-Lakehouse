# Spike Test: Spark to MinIO Connection
This notebook verifies that `spark` can read and write to the MinIO object storage using `delta` format.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("smoke-minio").getOrCreate()
df = spark.createDataFrame([(1, "ok")], ["id", "status"])
df.write.mode("overwrite").parquet("s3a://bronze/smoke-test/")
print("write_ok")
```
